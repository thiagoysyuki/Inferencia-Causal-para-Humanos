```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Generate sample data with correlation
x = np.linspace(0, 10, 50)
y_positive = 2 * x + np.random.normal(0, 2, 50)  # Positive correlation
y_negative = -2 * x + 20 + np.random.normal(0, 2, 50)  # Negative correlation
y_no_corr = np.random.normal(10, 5, 50)  # No correlation

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Positive correlation
axes[0].scatter(x, y_positive, alpha=0.6)
m1, b1 = np.polyfit(x, y_positive, 1)
axes[0].plot(x, m1*x + b1, color='blue', linewidth=2, label=f'r = {np.corrcoef(x, y_positive)[0, 1]:.3f}')
axes[0].set_title('Correlao Positiva')
axes[0].set_xlabel('Varivel X')
axes[0].set_ylabel('Varivel Y')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Negative correlation
axes[1].scatter(x, y_negative, alpha=0.6, color='red')
m2, b2 = np.polyfit(x, y_negative, 1)
axes[1].plot(x, m2*x + b2, color='darkred', linewidth=2, label=f'r = {np.corrcoef(x, y_negative)[0, 1]:.3f}')
axes[1].set_title('Correlao Negativa')
axes[1].set_xlabel('Varivel X')
axes[1].set_ylabel('Varivel Y')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# No correlation
axes[2].scatter(x, y_no_corr, alpha=0.6, color='green')
m3, b3 = np.polyfit(x, y_no_corr, 1)
axes[2].plot(x, m3*x + b3, color='darkgreen', linewidth=2, label=f'r = {np.corrcoef(x, y_no_corr)[0, 1]:.3f}')
axes[2].set_title('Sem Correlao')
axes[2].set_xlabel('Varivel X')
axes[2].set_ylabel('Varivel Y')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
